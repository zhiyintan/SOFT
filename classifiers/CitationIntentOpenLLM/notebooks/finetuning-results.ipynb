```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
base_data = {
    'Model': ['Instruct Q8', 'Instruct FP16', 'Finetuned Q8', 'Finetuned FP16'],
    'SciCite': [78.33, 80.41, 86.47, 86.84],
    'ACL-ARC': [61.04, 61.73, 68.48, 67.73]
}

base_df = pd.DataFrame(base_data)
print(base_df)

base_df.plot(x='Model', y=['SciCite', 'ACL-ARC'], kind='bar')
plt.ylabel('F1 Score')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.show()
```

```python
scicite_data = {
    'Model': ['Instruct Q8', 'Instruct FP16', 'Finetuned Q8', 'Finetuned FP16'],
    'Zero-Shot': [75.26, 75.38, 84.84, 84.49],
    'One-Shot': [76.29, 77.22, 85.46, 85.38],
    'Few-Shot': [78.33, 80.41, 86.47, 86.84],
    'Many-Shot': [78.32, 78.94, 85.62, 85.79]
}

scicite_df = pd.DataFrame(scicite_data)
display(scicite_df)

scicite_df.plot(x='Model', y=['Zero-Shot', 'One-Shot', 'Few-Shot', 'Many-Shot'], kind='bar', width=0.5)
plt.ylabel('F1 Score')
plt.xticks(rotation=45)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol=2)
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.show()
```
