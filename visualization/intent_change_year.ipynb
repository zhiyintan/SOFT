```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Construct the data directly into a list of tuples
df = pd.read_csv('intent_change_year.csv', sep='\t', encoding='utf-8')
citing_year = df['citing_year'].to_list()
citation_intent = df['citation_intent'].to_list()
citation_content = df['citation_content'].to_list()
```

```python
# Create a list of tuples for the records
records_intent = [(citing_year[i], citation_intent[i]) for i in range(len(citing_year))]

# Create DataFrame
df = pd.DataFrame(records_intent, columns=["citing_year", "citation_intent"])
for i in range(len(df)):
    if df.loc[i, 'citing_year'] < 1993:
        df.loc[i, 'citing_year'] = 1990
    if df.loc[i, 'citing_year'] > 1992 and df.loc[i, 'citing_year'] < 1998:
        df.loc[i, 'citing_year'] = 1995
    if df.loc[i, 'citing_year'] > 1997 and df.loc[i, 'citing_year'] < 2003:
        df.loc[i, 'citing_year'] = 2000
    if df.loc[i, 'citing_year'] > 2002 and df.loc[i, 'citing_year'] < 2008:
        df.loc[i, 'citing_year'] = 2005
    if df.loc[i, 'citing_year'] > 2007 and df.loc[i, 'citing_year'] < 2013:
        df.loc[i, 'citing_year'] = 2010
    if df.loc[i, 'citing_year'] > 2012 and df.loc[i, 'citing_year'] < 2018:
        df.loc[i, 'citing_year'] = 2015
    if df.loc[i, 'citing_year'] > 2017 and df.loc[i, 'citing_year'] < 2023:
        df.loc[i, 'citing_year'] = 2020
# Count occurrences per year per function
count_intent_df = df.groupby(['citing_year', 'citation_intent']).size().unstack(fill_value=0)
percent_intent_df = count_intent_df.div(count_intent_df.sum(axis=1), axis=0) * 100
```

```python
# Plotting the line chart
plt.figure(figsize=(12, 6))
percent_intent_df.plot(marker='o', figsize=(12, 6))
plt.title("Trends of Citation Functions over Time", fontsize=14, weight='bold')
plt.xlabel("Citing Year", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
